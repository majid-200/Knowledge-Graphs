```python
# This system builds a knowledge graph to recommend learning paths for educational content.
# It demonstrates how domain expertise can be encoded into a graph structure to help learners
# navigate complex topics in a structured way.

# Key Concepts:

# 1. NODES represent topics/concepts (e.g., "Python Basics", "Machine Learning")
# 2. EDGES represent relationships (e.g., "prerequisite", "related to")
# 3. WEIGHTS indicate relationship strength (0.0 = weak, 1.0 = strong)
# 4. METADATA adds context (difficulty, domain, keywords)

# Why Knowledge Graphs?

# - Visual representation of learning dependencies
# - Automated path finding for personalized learning
# - Identify key foundational topics (centrality analysis)
# - Scale to thousands of topics with minimal manual effort
```

# Define Topic Relationships (Edges)

```python
def define_relationships():
    """
    Define directed relationships between topics with associated weights.
    
    Each relationship is a tuple: (prerequisite_topic, dependent_topic, weight)
    
    Weight Interpretation:
    - 1.0: Strong prerequisite (must learn first topic before second)
    - 0.9: Highly recommended prerequisite
    - 0.8: Recommended prerequisite
    - 0.7: Helpful prerequisite (can learn concurrently)
    - 0.6: Loosely related (optional prerequisite)
    
    Returns:
        List of tuples: [(from_topic, to_topic, weight), ...]
    
    Graph Theory Note:
    This creates a Directed Acyclic Graph (DAG) where:
    - Direction indicates dependency (A → B means "learn A before B")
    - Weights indicate strength of dependency
    - No cycles (you can't have circular prerequisites)
    """
    
    return [
        # Python Core Learning Path
        # These form the foundation for other domains
        ('python_basics', 'data_types_variables', 1.0),  # Essential progression
        ('data_types_variables', 'control_flow', 1.0),    # Sequential learning
        ('control_flow', 'functions_basics', 1.0),        # Build on fundamentals
        ('functions_basics', 'data_structures', 0.9),     # Natural next step
        ('data_structures', 'file_handling', 0.8),        # Practical application
        ('data_structures', 'error_handling', 0.8),       # Best practices
        ('data_structures', 'oop_basics', 0.8),           # Advanced concepts
        ('oop_basics', 'advanced_oop', 0.7),              # Specialization
        
        # Finance Domain Path
        # Combines finance concepts with Python programming
        ('economics_101', 'stock_market_basics', 1.0),               # Foundation
        ('stock_market_basics', 'financial_statements', 0.9),        # Analysis prep
        ('financial_statements', 'financial_analysis_basics', 0.9),  # Core skills
        ('stock_market_basics', 'technical_analysis', 0.8),          # Parallel path
        
        # Cross-domain connections (Python + Finance)
        ('python_basics', 'python_finance', 0.7),                    # Programming foundation
        ('financial_analysis_basics', 'python_finance', 0.8),        # Domain knowledge
        ('data_structures', 'python_finance', 0.7),                  # Technical skills
        ('technical_analysis', 'python_finance', 0.7),               # Applied analysis
        ('python_finance', 'algorithmic_trading', 0.6),              # Advanced application
        
        # Bioimage Analysis Path
        # Combines biology/imaging with Python
        ('cell_biology_basics', 'microscopy_basics', 1.0),     # Biology foundation
        ('microscopy_basics', 'image_formats', 0.9),           # Technical understanding
        ('image_formats', 'image_analysis_basics', 0.9),       # Analysis preparation
        ('image_analysis_basics', 'image_processing', 0.8),    # Practical techniques
        
        # Cross-domain connections (Python + Bioimage)
        ('python_basics', 'python_imaging', 0.7),              # Programming foundation
        ('image_processing', 'python_imaging', 0.8),           # Domain knowledge
        ('data_structures', 'python_imaging', 0.7),            # Technical skills
        ('python_imaging', 'advanced_bioimage', 0.6),          # Advanced application
    ]
```

# Build the Knowledge Graph

```python
def build_graph():
    """
    Construct the NetworkX directed graph with all topics and relationships.
    
    Process:
    1. Create an empty directed graph
    2. Add all topic nodes with their metadata as node attributes
    3. Add all relationship edges with weights as edge attributes
    
    Returns:
        graph: NetworkX DiGraph object
        all_topics: Combined dictionary of all topic metadata
    
    Why DiGraph (Directed Graph)?
    - Direction matters: A→B means "A is prerequisite for B"
    - Allows us to traverse forward (what comes next?) or backward (what do I need first?)
    - Enables algorithms like shortest path, topological sort, etc.
    """
    
    # Initialize directed graph (edges have direction: A → B)
    graph = nx.DiGraph()
    
    # Get topic metadata from our definition function
    python_topics, finance_topics, bioimage_topics = create_topic_dictionaries()
    
    # Combine all topic dictionaries into one master dictionary
    # The ** operator unpacks dictionaries for merging
    all_topics = {**python_topics, **finance_topics, **bioimage_topics}
    
    # Add Nodes to Graph
    # Each topic becomes a node, with metadata stored as node attributes
    # Node attributes can be accessed later: graph.nodes[node_id]['difficulty']
    for topic_id, metadata in all_topics.items():
        graph.add_node(
            topic_id,  # Unique identifier for the node
            **metadata  # Unpack all metadata as node attributes
        )
    
    # Add Edges to Graph
    # Each relationship becomes a directed edge with weight and type
    relationships = define_relationships()
    for from_topic, to_topic, weight in relationships:
        graph.add_edge(
            from_topic,    # Source node (prerequisite)
            to_topic,      # Target node (dependent topic)
            weight=weight,  # Strength of relationship (0.0 to 1.0)
            type='prerequisite'  # Edge type (could have multiple types)
        )
    
    return graph, all_topics
```

```python
print("\nBuilding knowledge graph...")
graph, topic_metadata = build_graph()
print(f" Graph created with {graph.number_of_nodes()} topics")
print(f" Graph created with {graph.number_of_edges()} relationships")
```

# Interactive Visualization (Pyvis)

```python
def visualize_graph_pyvis(graph: nx.DiGraph, topic_metadata: Dict, 
                          output_file='learning_paths.html'):
    """
    Create an interactive HTML visualization of the knowledge graph.
    
    Features:
    - Color-coded nodes by domain
    - Hover tooltips showing topic details
    - Physics simulation for automatic layout
    - Interactive: drag, zoom, click nodes
    
    Args:
        graph: NetworkX graph to visualize
        topic_metadata: Dictionary of topic metadata
        output_file: Output HTML filename
    
    The Physics Engine:
    - Nodes repel each other (prevents overlap)
    - Edges act like springs (connected nodes stay close)
    - System reaches equilibrium for readable layout
    """
    
    # Initialize Pyvis network with configuration
    net = Network(
        notebook=True,      # Optimize for Jupyter notebooks
        height="750px",     # Canvas height
        width="100%",       # Full width
        bgcolor="#ffffff",  # White background
        directed=True       # Show arrow directions
    )
    
    # Color scheme for different learning domains
    # Using web-safe colors for consistent rendering
    color_map = {
        'programming': '#add8e6',  # Light blue - cool, logical
        'finance': '#90EE90',       # Light green - money/growth
        'biology': '#FA8072'        # Salmon - organic/life
    }
    
    # Add Nodes with Visual Styling
    for node in graph.nodes():
        metadata = graph.nodes[node]
        
        # Create formatted label (replace underscores, title case)
        label = node.replace('_', ' ').title()
        
        # Hover tooltip with detailed information
        tooltip = (
            f"Domain: {metadata['domain']}\n"
            f"Difficulty: {metadata['difficulty']}\n"
            f"Description: {metadata['description']}"
        )
        
        net.add_node(
            node,                                # Node ID
            label=label,                         # Display name
            title=tooltip,                       # Hover text
            color=color_map[metadata['domain']]  # Domain-based color
        )
    
    # Add Edges with Weights
    for edge in graph.edges(data=True):
        source, target, attributes = edge
        net.add_edge(
            source, 
            target, 
            title=f"Weight: {attributes['weight']}"  # Show weight on hover
        )
    
    # Generate and Save Visualization
    try:
        net.show(output_file)
        print(f"\nGraph visualization saved to {output_file}")
        print(f"Open this file in a web browser to explore the graph interactively")
    except Exception as e:
        print(f"Error in Pyvis visualization: {e}")
        print("Falling back to matplotlib static visualization...")
        visualize_graph_matplotlib(graph)
```

```python
visualize_graph_pyvis(graph, topic_metadata)
```

# Static Visualization (Matplotlib)

```python
def visualize_graph_matplotlib(graph: nx.DiGraph):
    """
    Create a static visualization using matplotlib as fallback.
    
    Best for:
    - Quick previews
    - Printed documents
    - Simple presentations
    
    Layout Algorithm:
    - Spring layout (Fruchterman-Reingold)
    - Treats edges as springs
    - Iteratively minimizes energy
    """
    
    plt.figure(figsize=(15, 10))
    
    # Calculate node positions using spring layout
    # k: optimal distance between nodes
    # iterations: number of optimization steps
    pos = nx.spring_layout(graph, k=1, iterations=50)
    
    # Assign Colors Based on Domain
    colors = []
    for node in graph.nodes():
        domain = graph.nodes[node]['domain']
        if domain == 'programming':
            colors.append('lightblue')
        elif domain == 'finance':
            colors.append('lightgreen')
        else:  # biology
            colors.append('salmon')
    
    # Draw the Graph
    nx.draw(
        graph, pos,
        node_color=colors,    # Domain-based colors
        with_labels=True,     # Show node names
        node_size=2000,       # Node circle size
        font_size=8,          # Label text size
        font_weight='bold',   # Make labels readable
        arrows=True,          # Show edge direction
        edge_color='gray',    # Subtle edge color
        arrowsize=20          # Arrow head size
    )
    
    plt.title("Learning Paths Knowledge Graph", fontsize=16)
    plt.tight_layout()
    plt.show()
```

# Graph Analysis Algorithms

```python
def basic_analysis(graph: nx.DiGraph):
    """
    Perform network analysis to identify important topics and patterns.
    
    Three Key Metrics:
    
    1. PageRank: Measures overall importance
       - Originally used by Google to rank web pages
       - High PageRank = many incoming links + links from important pages
       - In our context: Topics that many other topics depend on
    
    2. In-Degree Centrality: Direct prerequisite importance
       - Counts incoming edges
       - High in-degree = many topics require this as prerequisite
       - Foundation topics have high in-degree
    
    3. Betweenness Centrality: Bridge topics
       - Measures how often a node appears on shortest paths
       - High betweenness = topic connects different learning domains
       - Example: "Python basics" bridges all domains
    """
    
    
    # PageRank Analysis
    # Iterative algorithm that distributes "importance" through the graph
    pagerank = nx.pagerank(graph)
    
    print("\n" + "="*70)
    print("PAGERANK ANALYSIS - Overall Topic Importance")
    print("="*70)
    print("Higher scores indicate more central/important topics in the curriculum")
    print("-"*70)
    
    for node, rank in sorted(pagerank.items(), key=lambda x: x[1], reverse=True):
        # Format node name for display
        display_name = node.replace('_', ' ').title()
        # Create bar visualization
        bar = '█' * int(rank * 500)
        print(f"{display_name:35} {rank:.4f} {bar}")
        
    
    # In-Degree Centrality Analysis
    # Normalized count of incoming edges (prerequisites)
    in_degree = nx.in_degree_centrality(graph)
    
    print("\n" + "="*70)
    print("IN-DEGREE CENTRALITY - Critical Prerequisites")
    print("="*70)
    print("Topics that many other topics depend on")
    print("-"*70)
    
    for node, centrality in sorted(in_degree.items(), key=lambda x: x[1], reverse=True)[:10]:
        if centrality > 0:  # Only show topics with dependencies
            display_name = node.replace('_', ' ').title()
            bar = '█' * int(centrality * 100)
            print(f"{display_name:35} {centrality:.4f} {bar}")
    

    # Betweenness Centrality Analysis
    # Measures how often nodes appear on shortest paths between other nodes
    betweenness = nx.betweenness_centrality(graph)
    
    print("\n" + "="*70)
    print("BETWEENNESS CENTRALITY - Bridge Topics")
    print("="*70)
    print("Topics that connect different learning domains")
    print("-"*70)
    
    for node, centrality in sorted(betweenness.items(), key=lambda x: x[1], reverse=True)[:10]:
        if centrality > 0:  # Only show bridges
            display_name = node.replace('_', ' ').title()
            bar = '█' * int(centrality * 200)
            print(f"{display_name:35} {centrality:.4f} {bar}")
```

# Prerequisite Discovery

```python
def get_prerequisites(graph: nx.DiGraph, topic: str, max_depth=3):
    """
    Find all prerequisites for a topic using breadth-first search (BFS).
    
    Algorithm:
    1. Start at the target topic
    2. Find all direct prerequisites (incoming edges)
    3. For each prerequisite, find its prerequisites
    4. Continue until max_depth reached or no more prerequisites
    
    Args:
        graph: Knowledge graph
        topic: Target topic to analyze
        max_depth: How many levels deep to search (prevents infinite loops)
    
    Returns:
        Dictionary mapping prerequisite topics to their depth level
        Example: {'python_basics': 1, 'data_types': 2}
    
    Why BFS instead of DFS?
    - BFS finds shortest path first (optimal learning path)
    - DFS might explore unnecessarily deep paths
    """
    
    # Validate input
    if topic not in graph.nodes():
        print(f"Error: Topic '{topic}' not found in the graph")
        return {}
    
    # Initialize tracking structures
    prerequisites = {}      # Store results: {topic: depth}
    current_depth = 1       # Start at depth 1 (immediate prerequisites)
    current_nodes = {topic} # Nodes to explore at current level
    visited = set()         # Prevent revisiting nodes
    

    # Breadth-First Search (BFS) Algorithm
    while current_depth <= max_depth and current_nodes:
        next_nodes = set()  # Nodes to explore at next level
        
        # Process all nodes at current depth level
        for node in current_nodes:
            # Get all incoming edges (prerequisites)
            for predecessor in graph.predecessors(node):
                # Only add if not already visited
                if predecessor not in visited and predecessor not in prerequisites:
                    prerequisites[predecessor] = current_depth
                    next_nodes.add(predecessor)
            
            visited.add(node)
        
        # Move to next depth level
        current_nodes = next_nodes
        current_depth += 1
    
    return prerequisites


def print_prerequisites_analysis(graph: nx.DiGraph, topic: str):
    """
    Display prerequisite analysis in a formatted, hierarchical view.
    
    Shows prerequisites organized by "distance" from target topic:
    - Level 1: Direct prerequisites
    - Level 2: Prerequisites of prerequisites
    - Level 3: Foundation topics
    """
    
    print(f"\n{'='*70}")
    print(f"PREREQUISITE ANALYSIS FOR: {topic.replace('_', ' ').title()}")
    print(f"{'='*70}")
    
    # Get prerequisites using BFS
    prereqs = get_prerequisites(graph, topic)
    
    if not prereqs:
        print("No prerequisites found or topic is foundational.")
        return
    
    # Group prerequisites by their depth level
    depth_groups = defaultdict(list)
    for prereq_topic, depth in prereqs.items():
        depth_groups[depth].append(prereq_topic)

    
    # Display Prerequisites Hierarchically
    for depth in sorted(depth_groups.keys()):
        print(f"\n{'-'*70}")
        print(f"LEVEL {depth} PREREQUISITES (Learn these {'first' if depth == 1 else f'{depth} steps before'})")
        print(f"{'-'*70}")
        
        for prereq in sorted(depth_groups[depth]):
            metadata = graph.nodes[prereq]
            display_name = prereq.replace('_', ' ').title()
            
            # Format with indentation and metadata
            print(f"\n{display_name}")
            print(f"     Difficulty: {metadata['difficulty'].upper()}")
            print(f"     Domain: {metadata['domain'].title()}")
            print(f"     Description: {metadata['description']}")
```

```python
# Example 1: Prerequisites for Algorithmic Trading
print_prerequisites_analysis(graph, 'algorithmic_trading')

# Example 2: Prerequisites for Advanced Bioimage Analysis
print_prerequisites_analysis(graph, 'advanced_bioimage')
```

# Learning Path Recommendation

```python
def recommend_learning_path(graph: nx.DiGraph, domain: str):
    """
    Generate an optimal learning path for a specific domain.
    
    Uses Dijkstra's shortest path algorithm to find the most efficient
    route from beginner topics to advanced topics in the domain.
    
    Args:
        graph: Knowledge graph
        domain: Target domain ('finance' or 'bioimage')
    
    Algorithm:
    1. Identify starting point (python_basics)
    2. Identify goal (advanced topic in domain)
    3. Find shortest path considering edge weights
    4. Display path with difficulty transitions
    5. Suggest supplementary topics
    """
    
    # Validate domain input
    if domain not in ['finance', 'bioimage']:
        print("Error: Domain must be either 'finance' or 'bioimage'")
        return
    
    print(f"\n{'='*70}")
    print(f"RECOMMENDED LEARNING PATH: Python for {domain.upper()}")
    print(f"{'='*70}")
    

    # Define Domain Parameters
    if domain == 'finance':
        target_node = 'algorithmic_trading'  # End goal
        domain_identifier = 'finance'         # For filtering
    else:  # bioimage
        target_node = 'advanced_bioimage'
        domain_identifier = 'biology'
    
    
    # Find Optimal Path Using NetworkX
    try:
        # Find ALL shortest paths (there might be multiple equally good paths)
        # weight=None means we treat all edges equally (unweighted shortest path)
        paths = nx.all_shortest_paths(
            graph, 
            source='python_basics',  # Start with programming fundamentals
            target=target_node,       # End at advanced domain topic
            weight=None               # Could use 'weight' for weighted shortest path
        )
        paths = list(paths)  # Convert generator to list
        
    except nx.NetworkXNoPath:
        print(f"No learning path found from Python basics to {domain}")
        return
    
    # Select the first shortest path (could add logic to pick "best" path)
    optimal_path = paths[0]
    

    # Display Learning Sequence
    print("\n📖 Recommended Learning Sequence:")
    print("-"*70)
    
    for i, current_topic in enumerate(optimal_path, 1):
        metadata = graph.nodes[current_topic]
        display_name = current_topic.replace('_', ' ').title()
        
        # Create difficulty badge
        difficulty_badge = {
            'beginner': '🟢',
            'intermediate': '🟡',
            'advanced': '🔴'
        }.get(metadata['difficulty'], '⚪')
        
        print(f"\n{i}. {display_name}")
        print(f"   {difficulty_badge} Difficulty: {metadata['difficulty'].upper()}")
        print(f"   📋 {metadata['description']}")
        

        # Calculate Transition Difficulty
        # Show how hard it is to move from this topic to the next
        if i < len(optimal_path):
            next_topic = optimal_path[i]
            edge_data = graph.get_edge_data(current_topic, next_topic)
            
            if edge_data and 'weight' in edge_data:
                # Lower weight = harder transition
                transition_difficulty = (1 - edge_data['weight']) * 100
                
                # Create visual indicator
                if transition_difficulty < 20:
                    indicator = "✅ Easy transition"
                elif transition_difficulty < 40:
                    indicator = "⚠️  Moderate jump"
                else:
                    indicator = "🔥 Challenging jump"
                
                print(f"   {indicator} (difficulty: {transition_difficulty:.0f}%)")
    

    # Suggest Supplementary Topics
    print(f"\n{'='*70}")
    print("💡 SUPPLEMENTARY TOPICS (Optional but helpful)")
    print("="*70)
    print("These topics aren't on the critical path but provide valuable context:\n")
    
    for node, metadata in graph.nodes(data=True):
        # Find topics in the same domain not on the main path
        if (metadata['domain'] == domain_identifier and 
            node not in optimal_path and 
            metadata['difficulty'] != 'advanced'):  # Exclude advanced topics
            
            display_name = node.replace('_', ' ').title()
            difficulty_badge = '🟢' if metadata['difficulty'] == 'beginner' else '🟡'
            print(f"  {difficulty_badge} {display_name} ({metadata['difficulty']})")
```

# Utility Functions for Extended Functionality

```python
def find_topics_by_keyword(graph: nx.DiGraph, keyword: str):
    """
    Search for topics matching a specific keyword.
    
    Useful for:
    - Content discovery
    - Building dynamic curricula
    - Matching learner interests to topics
    
    Args:
        graph: Knowledge graph
        keyword: Search term (case-insensitive)
    
    Returns:
        List of matching topics with metadata
    """
    keyword_lower = keyword.lower()
    matches = []
    
    for node, data in graph.nodes(data=True):
        # Search in keywords list and description
        if (keyword_lower in [k.lower() for k in data['keywords']] or
            keyword_lower in data['description'].lower()):
            matches.append({
                'topic': node,
                'domain': data['domain'],
                'difficulty': data['difficulty'],
                'description': data['description']
            })
    
    return matches


def get_learning_path_by_difficulty(graph: nx.DiGraph, domain: str, 
                                    start_difficulty: str = 'beginner'):
    """
    Generate a learning path filtered by starting difficulty level.
    
    Use case: Learners with existing knowledge can skip beginner content
    
    Args:
        graph: Knowledge graph
        domain: Target domain
        start_difficulty: Minimum difficulty level to include
    
    Returns:
        Filtered learning path
    """
    difficulty_levels = {'beginner': 0, 'intermediate': 1, 'advanced': 2}
    min_level = difficulty_levels.get(start_difficulty, 0)
    
    # Filter nodes by domain and difficulty
    valid_nodes = [
        node for node, data in graph.nodes(data=True)
        if (data['domain'] == domain or data['domain'] == 'programming') and
           difficulty_levels.get(data['difficulty'], 0) >= min_level
    ]
    
    # Create subgraph with only valid nodes
    subgraph = graph.subgraph(valid_nodes)
    
    return subgraph


def calculate_learning_time(graph: nx.DiGraph, path: List[str], 
                           time_per_topic: Dict[str, int] = None):
    """
    Estimate total learning time for a path.
    
    Args:
        graph: Knowledge graph
        path: List of topics in learning sequence
        time_per_topic: Custom time estimates (hours) per topic
    
    Returns:
        Total estimated hours
    """
    # Default time estimates by difficulty
    default_times = {
        'beginner': 5,      # 5 hours per beginner topic
        'intermediate': 10,  # 10 hours per intermediate topic
        'advanced': 20       # 20 hours per advanced topic
    }
    
    total_time = 0
    
    for topic in path:
        if time_per_topic and topic in time_per_topic:
            total_time += time_per_topic[topic]
        else:
            difficulty = graph.nodes[topic]['difficulty']
            total_time += default_times[difficulty]
    
    return total_time


def export_to_json(graph: nx.DiGraph, filename: str = 'knowledge_graph.json'):
    """
    Export graph to JSON for external use (web apps, APIs, etc.).
    
    Format:
    {
        "nodes": [{"id": "...", "difficulty": "...", ...}],
        "edges": [{"source": "...", "target": "...", "weight": ...}]
    }
    """
    data = {
        'nodes': [
            {'id': node, **graph.nodes[node]}
            for node in graph.nodes()
        ],
        'edges': [
            {
                'source': edge[0],
                'target': edge[1],
                'weight': edge[2].get('weight', 1.0),
                'type': edge[2].get('type', 'prerequisite')
            }
            for edge in graph.edges(data=True)
        ]
    }
    
    with open(filename, 'w') as f:
        json.dump(data, f, indent=2)
    
    print(f"✓ Graph exported to {filename}")


def save_to_database(graph: nx.DiGraph, db_path: str = 'knowledge_graph.db'):
    """
    Store graph in SQLite database for persistence and querying.
    
    Creates two tables:
    - topics: All topic metadata
    - relationships: All edges with weights
    """
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    
    # Create topics table
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS topics (
            topic_id TEXT PRIMARY KEY,
            difficulty TEXT,
            domain TEXT,
            description TEXT,
            keywords TEXT
        )
    ''')
    
    # Create relationships table
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS relationships (
            from_topic TEXT,
            to_topic TEXT,
            weight REAL,
            type TEXT,
            FOREIGN KEY (from_topic) REFERENCES topics(topic_id),
            FOREIGN KEY (to_topic) REFERENCES topics(topic_id)
        )
    ''')
    
    # Insert topics
    for node, data in graph.nodes(data=True):
        cursor.execute(
            'INSERT OR REPLACE INTO topics VALUES (?, ?, ?, ?, ?)',
            (node, data['difficulty'], data['domain'], 
             data['description'], json.dumps(data['keywords']))
        )
    
    # Insert relationships
    for edge in graph.edges(data=True):
        cursor.execute(
            'INSERT OR REPLACE INTO relationships VALUES (?, ?, ?, ?)',
            (edge[0], edge[1], edge[2]['weight'], edge[2]['type'])
        )
    
    conn.commit()
    conn.close()
    print(f"Graph saved to database: {db_path}")
```
