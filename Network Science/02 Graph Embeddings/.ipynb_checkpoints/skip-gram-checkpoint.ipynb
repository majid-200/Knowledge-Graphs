# Node Embeddings with GNNs - Skip-Gram

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import networkx as nx
import numpy as np
import random
from gensim.models.word2vec import Word2Vec
from sklearn.manifold import TSNE
```

## UNDERSTANDING WORD2VEC (Foundation for Graph Embeddings)

```python
# Word2Vec is a neural network model that learns word embeddings from text.
# It uses the Skip-Gram architecture which predicts context words given a target word.

# Key Idea: Words that appear in similar contexts should have similar embeddings.
# For graphs: Nodes that appear in similar random walks should have similar embeddings.

# Skip-Gram Architecture:
# Input: Target word (one-hot encoded)
# Hidden Layer: Embedding layer (word → vector)
# Output: Context words (predicts surrounding words)
```

## RandomWalk

```python
# Random walks are the bridge between graphs and text processing.

# Process:
# 1. Start at a node
# 2. Randomly move to a neighbor
# 3. Repeat for a fixed number of steps
# 4. The sequence of visited nodes forms a "sentence"

# These "sentences" can then be fed into Word2Vec to learn node embeddings.
```

```python
random.seed(0)

class RandomWalk:
    """
    Performs unbiased random walks on a graph.
    This is the foundation for DeepWalk.
    """
    
    def __init__(self, graph):
        """
        Initialize random walker.
        
        Parameters:
        -----------
        graph : networkx.Graph
            Input graph for random walks
        """
        self.graph = graph
    
    def random_walk(self, start, length):
        """
        Perform a single random walk starting from a node.
        
        Parameters:
        -----------
        start : int
            Starting node ID
        length : int
            Number of steps in the walk
            
        Returns:
        --------
        list : Sequence of visited nodes (as strings)
        """
        walk = [str(start)]  # Initialize walk with starting node
        
        # Perform random walk steps
        for i in range(length):
            # Get neighbors of current node
            neighbors = list(self.graph.neighbors(start))
            
            if len(neighbors) == 0:
                break  # Dead end - stop walking
            
            # Randomly select next node (uniform probability)
            next_node = np.random.choice(neighbors, 1)[0]
            walk.append(str(next_node))
            start = next_node
        
        return walk
    
    def plot_graph(self):
        """Visualize the graph structure"""
        plt.figure(figsize=(8, 6), dpi=150)
        plt.axis('off')
        
        # Use spring layout for better visualization
        pos = nx.spring_layout(self.graph, seed=0)
        
        nx.draw_networkx(
            self.graph,
            pos=pos,
            node_size=600,
            node_color='lightblue',
            cmap='coolwarm',
            font_size=14,
            font_color='white',
            edge_color='gray',
            alpha=0.8
        )
        plt.title('Graph Structure', fontweight='bold', fontsize=14)
        plt.tight_layout()
        plt.show()
```

```python
# Demonstration: Random Walk on Erdős-Rényi Graph
print("\nGenerating random graph...")
G = nx.erdos_renyi_graph(n=10, p=0.3, seed=1, directed=False)

print(f"Graph properties:")
print(f"  - Nodes: {G.number_of_nodes()}")
print(f"  - Edges: {G.number_of_edges()}")
print(f"  - Density: {nx.density(G):.3f}")

# Perform a random walk
random_walker = RandomWalk(G)
walk_result = random_walker.random_walk(start=0, length=10)

print(f"\nRandom walk starting from node 0:")
print(f"  Path: {' → '.join(walk_result)}")
print(f"  Length: {len(walk_result)} nodes visited")

# Visualize the graph
random_walker.plot_graph()
```

## DeepWalk

```python
# DeepWalk Algorithm:
# 1. Generate multiple random walks from each node
# 2. Treat each walk as a "sentence"
# 3. Apply Word2Vec Skip-Gram to learn embeddings
# 4. Nodes appearing in similar contexts get similar embeddings

# Original Paper: "DeepWalk: Online Learning of Social Representations" (KDD 2014)
```

```python
random.seed(0)

class DeepWalk(RandomWalk):
    """
    DeepWalk: Learn node embeddings using random walks and Word2Vec.
    Extends RandomWalk class to generate multiple walks and train embeddings.
    """
    
    def __init__(self, graph):
        """Initialize DeepWalk with a graph"""
        super().__init__(graph)
    
    def generate_random_walks(self, num_walks, walk_length):
        """
        Generate multiple random walks from each node in the graph.
        
        Parameters:
        -----------
        num_walks : int
            Number of walks to generate per node
        walk_length : int
            Length of each walk
            
        Returns:
        --------
        list : Collection of all random walks (sentences)
        """
        walks = []
        
        print(f"Generating {num_walks} walks of length {walk_length} from each node...")
        
        # Generate walks from each node
        for node in self.graph.nodes():
            for walk_num in range(num_walks):
                walk = self.random_walk(node, walk_length)
                walks.append(walk)
        
        print(f"✓ Generated {len(walks)} total walks")
        return walks
    
    def train_word2vec(self, walks, vector_size=100, window=10, epochs=30):
        """
        Train Word2Vec model on random walks to learn node embeddings.
        
        Parameters:
        -----------
        walks : list
            Collection of random walks
        vector_size : int
            Embedding dimension
        window : int
            Context window size for Skip-Gram
        epochs : int
            Number of training epochs
            
        Returns:
        --------
        Word2Vec model with learned node embeddings
        """
        print(f"\nTraining Word2Vec on {len(walks)} walks...")
        print(f"  - Embedding dimension: {vector_size}")
        print(f"  - Window size: {window}")
        print(f"  - Epochs: {epochs}")
        
        # Initialize Word2Vec with Skip-Gram
        model = Word2Vec(
            walks,
            hs=1,                   # Hierarchical softmax (faster training)
            sg=1,                   # Skip-gram architecture
            vector_size=vector_size,
            window=window,
            workers=1,
            seed=1
        )
        
        # Build vocabulary from walks
        model.build_vocab(walks)
        
        # Train the model
        model.train(
            walks, 
            total_examples=model.corpus_count, 
            epochs=epochs,
            report_delay=1
        )
        
        print("✓ Training completed!")
        return model
    
    def visualize_tsne(self, model, labels):
        """
        Visualize high-dimensional embeddings in 2D using t-SNE.
        
        Parameters:
        -----------
        model : Word2Vec
            Trained Word2Vec model
        labels : list
            Node labels for coloring (e.g., community assignments)
        """
        print("\nApplying t-SNE for visualization...")
        
        # Extract node embeddings in order
        nodes_wv = np.array([
            model.wv.get_vector(str(i)) 
            for i in range(len(model.wv))
        ])
        
        # Apply t-SNE dimensionality reduction
        tsne = TSNE(
            n_components=2,         # Reduce to 2D
            learning_rate='auto',
            init='pca',             # Initialize with PCA
            random_state=0
        ).fit_transform(nodes_wv)
        
        # Plot results
        plt.figure(figsize=(8, 8))
        scatter = plt.scatter(
            tsne[:, 0], 
            tsne[:, 1], 
            s=200, 
            c=labels, 
            cmap="coolwarm",
            edgecolors='black',
            linewidth=1.5,
            alpha=0.8
        )
        plt.colorbar(scatter, label='Community')
        plt.title('DeepWalk Embeddings (t-SNE Visualization)', 
                  fontweight='bold', fontsize=14)
        plt.xlabel('t-SNE Dimension 1')
        plt.ylabel('t-SNE Dimension 2')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
```

```python
# ============================================================================
# DEEPWALK DEMONSTRATION: KARATE CLUB GRAPH
# ============================================================================

print("\n" + "-"*70)
print("DeepWalk on Zachary's Karate Club")
print("-"*70)

# Load famous Karate Club graph
G = nx.karate_club_graph()

print(f"\nKarate Club Graph:")
print(f"  - Nodes: {G.number_of_nodes()} (members)")
print(f"  - Edges: {G.number_of_edges()} (friendships)")
print(f"  - Two factions: Mr. Hi vs Officer")

# Extract community labels (0 = Mr. Hi, 1 = Officer)
labels = [1 if G.nodes[node]['club'] == 'Officer' else 0 for node in G.nodes()]

print(f"  - Mr. Hi faction: {labels.count(0)} members")
print(f"  - Officer faction: {labels.count(1)} members")

# Create DeepWalk instance
deep_walker = DeepWalk(G)

# Generate random walks
walks = deep_walker.generate_random_walks(num_walks=100, walk_length=10)

# Train Word2Vec model
word2vec_model = deep_walker.train_word2vec(
    walks, 
    vector_size=100,    # High-dimensional embeddings
    window=10, 
    epochs=30
)

# Visualize embeddings using t-SNE
deep_walker.visualize_tsne(word2vec_model, labels)
```

```python
# ============================================================================
# DEEPWALK APPLICATION: FINDING SIMILAR NODES
# ============================================================================

print("\n" + "-"*70)
print("Finding Similar Nodes")
print("-"*70)

print('\nNodes most similar to node 0 (based on embeddings):')
for similarity in word2vec_model.wv.most_similar(positive=['0'], topn=5):
    node_id, score = similarity
    print(f'  Node {node_id}: similarity = {score:.4f}')
```

## Node2Vec (BIASED RANDOM WALKS)

```python
# Node2Vec improves upon DeepWalk by using BIASED random walks.

# Key Innovation: Two parameters control exploration strategy:
# - p (return parameter): Controls likelihood of revisiting previous nodes
#   * p > 1: Less likely to return (explore more)
#   * p < 1: More likely to return (stay local)
  
# - q (in-out parameter): Controls exploration vs exploitation
#   * q > 1: Stay close (BFS-like, local structure)
#   * q < 1: Explore far (DFS-like, global structure)

# This allows Node2Vec to balance between:
# - Homophily: Nearby nodes should be similar
# - Structural equivalence: Nodes with similar roles should be similar

# Original Paper: "node2vec: Scalable Feature Learning for Networks" (KDD 2016)
```

```python
class Node2Vec:
    """
    Node2Vec: Learn node embeddings using biased random walks.
    Uses parameters p and q to control exploration strategy.
    """
    
    def __init__(self, graph, dimensions=128, walk_length=30, 
                 num_walks=10, p=1.0, q=1.0):
        """
        Initialize Node2Vec.
        
        Parameters:
        -----------
        graph : networkx.Graph
            Input graph
        dimensions : int
            Embedding dimension
        walk_length : int
            Length of each random walk
        num_walks : int
            Number of walks per node
        p : float
            Return parameter (controls backtracking)
        q : float
            In-out parameter (controls exploration)
        """
        self.graph = graph
        self.dimensions = dimensions
        self.walk_length = walk_length
        self.num_walks = num_walks
        self.p = p
        self.q = q
        self.embeddings = None
    
    def _biased_random_walk(self, start_node):
        """
        Perform a single biased random walk using p and q parameters.
        
        The bias determines the probability of:
        - Returning to previous node (controlled by p)
        - Staying local vs exploring (controlled by q)
        """
        walk = [start_node]
        
        for _ in range(self.walk_length - 1):
            current = walk[-1]
            neighbors = list(self.graph.neighbors(current))
            
            if len(neighbors) == 0:
                break
            
            if len(walk) == 1:
                # First step: uniform random choice
                walk.append(np.random.choice(neighbors))
            else:
                # Biased walk based on previous node
                prev = walk[-2]
                
                # Calculate transition probabilities
                weights = []
                for neighbor in neighbors:
                    if neighbor == prev:
                        # Return to previous node: weight = 1/p
                        weights.append(1.0 / self.p)
                    elif self.graph.has_edge(neighbor, prev):
                        # Stay local: weight = 1
                        weights.append(1.0)
                    else:
                        # Explore outward: weight = 1/q
                        weights.append(1.0 / self.q)
                
                # Normalize probabilities
                probabilities = np.array(weights) / np.sum(weights)
                walk.append(np.random.choice(neighbors, p=probabilities))
        
        return [str(node) for node in walk]
    
    def generate_walks(self):
        """Generate all biased random walks for training"""
        walks = []
        print(f"Generating {self.num_walks} biased walks per node (p={self.p}, q={self.q})...")
        
        for _ in range(self.num_walks):
            for node in self.graph.nodes():
                walks.append(self._biased_random_walk(node))
        
        print(f"✓ Generated {len(walks)} walks")
        return walks
    
    def train(self, walks):
        """Train Word2Vec on biased walks"""
        print(f"\nTraining Word2Vec on Node2Vec walks...")
        
        word2vec = Word2Vec(
            walks, 
            vector_size=self.dimensions, 
            window=5, 
            min_count=0,
            sg=1,  # Skip-Gram
            workers=1,
            seed=42
        )
        
        # Store embeddings
        self.embeddings = {
            str(node): word2vec.wv[str(node)] 
            for node in self.graph.nodes()
        }
        
        print("✓ Training completed!")
    
    def visualize_embeddings(self, G):
        """Visualize embeddings using t-SNE"""
        if self.embeddings is None:
            print("ERROR: Embeddings not trained yet!")
            return
        
        # Get binary cluster labels
        labels = [1 if G.nodes[node]['club'] == 'Officer' else 0 for node in G.nodes()]
        
        node_ids, embeddings = zip(*self.embeddings.items())
        embeddings = np.array(embeddings)
        
        # Apply t-SNE
        print("\nApplying t-SNE...")
        tsne = TSNE(n_components=2, random_state=42)
        embeddings_2d = tsne.fit_transform(embeddings)
        
        # Visualize
        plt.figure(figsize=(10, 8))
        scatter = plt.scatter(
            embeddings_2d[:, 0], 
            embeddings_2d[:, 1], 
            c=labels, 
            cmap="coolwarm",
            s=200,
            edgecolors='black',
            linewidth=1.5,
            alpha=0.8
        )
        
        for i, node_id in enumerate(node_ids):
            plt.annotate(
                node_id, 
                (embeddings_2d[i, 0], embeddings_2d[i, 1]), 
                alpha=0.7,
                fontsize=8
            )
        
        plt.colorbar(scatter, label='Community')
        plt.title(f'Node2Vec Embeddings (p={self.p}, q={self.q})', 
                  fontweight='bold', fontsize=14)
        plt.xlabel('t-SNE Dimension 1')
        plt.ylabel('t-SNE Dimension 2')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
```

```python
# Load graph
G = nx.karate_club_graph()

# Initialize Node2Vec
node2vec = Node2Vec(
    G, 
    dimensions=128,     # High-dimensional embeddings
    walk_length=30, 
    num_walks=10, 
    p=1.0,             # Return parameter
    q=1.0              # In-out parameter
)

# Generate biased walks
walks = node2vec.generate_walks()

# Train embeddings
node2vec.train(walks)

# Visualize
node2vec.visualize_embeddings(G)
```

## Pre-built Node2Vec

```python
from node2vec import Node2Vec as Node2VecLib

# Initialize with 2D embeddings for visualization
print("Initializing Node2Vec...")
node2vec_model = Node2VecLib(
    G, 
    dimensions=2,       # 2D for visualization
    walk_length=30,
    num_walks=100,
    p=1,
    q=1,
    workers=1
)

print("Training model...")
model = node2vec_model.fit(window=10, min_count=1, batch_words=4)
print("✓ Training completed!")
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 7))

# Get labels
labels = [1 if G.nodes[node]['club'] == 'Officer' else 0 for node in G.nodes()]

# Left: Original graph
nx.draw_networkx(
    G,
    pos=nx.spring_layout(G, seed=0),
    node_color=labels,
    node_size=300,
    cmap='coolwarm',
    font_size=8,
    font_color='white',
    edge_color='gray',
    alpha=0.8,
    ax=axs[0]
)
axs[0].set_title('Karate Club Graph (Original)', fontweight='bold', fontsize=14)
axs[0].axis('off')

# Right: Node2Vec embeddings
labels_array = np.array(labels)
norm_labels = (labels_array - labels_array.min()) / (labels_array.max() - labels_array.min())
mapped_colors = cm.coolwarm(norm_labels)

for node_id in G.nodes():
    v = model.wv.get_vector(str(node_id))
    axs[1].scatter(
        v[0], v[1], 
        s=300, 
        c=[mapped_colors[node_id]],
        edgecolors='black',
        linewidth=1.5
    )
    axs[1].annotate(
        str(node_id), 
        (v[0], v[1]), 
        fontsize=8,
        ha='center',
        va='center',
        fontweight='bold'
    )

axs[1].set_title('Node2Vec Node Embeddings (2D)', fontweight='bold', fontsize=14)
axs[1].set_xlabel('Embedding Dimension 1', fontweight='bold')
axs[1].set_ylabel('Embedding Dimension 2', fontweight='bold')
axs[1].grid(True, alpha=0.3)

plt.subplots_adjust(wspace=0.3)
fig.suptitle('Node2Vec: Biased Random Walks → 2D Embeddings', 
             fontweight='bold', fontsize=16)
plt.tight_layout()
plt.show()
```
