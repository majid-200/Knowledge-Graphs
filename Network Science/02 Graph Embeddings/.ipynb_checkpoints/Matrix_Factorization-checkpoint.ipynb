## GraphFactorization 

```python
# Graph Factorization is a matrix factorization approach that learns node embeddings
# by approximating the adjacency matrix of the graph. It optimizes an objective function
# that minimizes the reconstruction error between the original and approximated graphs.

# Key idea: Represent each node as a low-dimensional vector such that the inner product
# of two node vectors approximates their edge weight in the original graph.

# L = (1/2) · Σ(i,j)∈E (Aij - Yi,: · Yj,:^T)² + (λ/2) · Σi ||Yi,:||²

# L = Loss function (what we want to minimize)
# Aij = Adjacency matrix element (1 if edge exists, 0 otherwise)
# Yi,: = Embedding vector for node i (row i of embedding matrix Y)
# Yj,:^T = Transpose of embedding vector for node j
# Yi,: · Yj,:^T = Dot product of embeddings (scalar value)
# E = Set of all edges in the graph
# (1/2) = Scaling factor (makes gradient calculations cleaner)
# λ = Regularization parameter (prevents overfitting)
# ||Yi,:||² = L2 norm squared = sum of squared elements of Yi,:

# Reconstruction Loss (first term):

# (1/2) · Σ(i,j)∈E (Aij - Yi,: · Yj,:^T)²
# Penalizes difference between actual edges and predicted edges
# Only computed for edges that exist in the graph


# Regularization Loss (second term):

# (λ/2) · Σi ||Yi,:||²
# Penalizes large embedding values
# Prevents overfitting and improves generalization
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

class GraphFactorization:
    """
    Graph Factorization for learning node embeddings.
    
    Minimizes: ||A - U * U^T||^2 + regu * ||U||^2
    where A is the adjacency matrix and U is the embedding matrix.
    
    Parameters:
    -----------
    d : int
        Embedding dimension (number of features per node)
    max_iter : int
        Maximum number of gradient descent iterations
    eta : float
        Learning rate for gradient descent
    regu : float
        Regularization parameter (L2 regularization strength)
    """
    
    def __init__(self, d=128, max_iter=10000, eta=1e-4, regu=1.0):
        self.d = d
        self.max_iter = max_iter
        self.eta = eta
        self.regu = regu
        self.embedding = None
        self._node_list = None
        
    def learn_embedding(self, graph):
        """
        Learn embeddings for nodes in the graph.
        
        Parameters:
        -----------
        graph : networkx.Graph
            Input graph
        """
        # Get adjacency matrix and node list
        self._node_list = list(graph.nodes())
        A = nx.to_numpy_array(graph, nodelist=self._node_list)
        n = A.shape[0]
        
        # Initialize embedding matrix with small random values
        np.random.seed(42)
        U = np.random.randn(n, self.d) * 0.01
        
        print(f"Starting optimization with {n} nodes...")
        
        # Gradient descent optimization
        for iteration in range(self.max_iter):
            # Compute reconstruction: A_hat = U * U^T
            A_hat = U @ U.T
            
            # Compute error
            diff = A - A_hat
            
            # Compute loss: reconstruction error + regularization
            reconstruction_loss = np.sum(diff ** 2)
            regularization_loss = self.regu * np.sum(U ** 2)
            total_loss = reconstruction_loss + regularization_loss
            
            # Compute gradient: dL/dU = -2 * (A - U*U^T) * U + 2 * regu * U
            grad = -2 * (diff @ U) + 2 * self.regu * U
            
            # Update embeddings
            U -= self.eta * grad
            
            # Print progress every 1000 iterations
            if (iteration + 1) % 1000 == 0:
                print(f"Iteration {iteration + 1}/{self.max_iter} - Loss: {total_loss:.4f} "
                      f"(Reconstruction: {reconstruction_loss:.4f}, Reg: {regularization_loss:.4f})")
        
        self.embedding = U
        print("Training complete!")
        
    def get_embedding(self):
        """
        Get the learned node embeddings.
        
        Returns:
        --------
        numpy.ndarray
            Node embedding matrix of shape (n_nodes, d)
        """
        return self.embedding
    
    def get_reconstructed_adj(self):
        """
        Get the reconstructed adjacency matrix A_hat = U * U^T
        
        Returns:
        --------
        numpy.ndarray
            Reconstructed adjacency matrix
        """
        if self.embedding is None:
            raise ValueError("Model has not been trained yet. Call learn_embedding() first.")
        return self.embedding @ self.embedding.T
```

```python
# ----------------------------------------------------------------------------
# Create Example Graph: Barbell Graph
# ----------------------------------------------------------------------------
# A barbell graph consists of two complete graphs (cliques) connected by a path
# m1: number of nodes in each complete graph (clique)
# m2: number of nodes in the connecting path
G = nx.barbell_graph(m1=10, m2=4)

print(f"  - Number of nodes: {G.number_of_nodes()}")
print(f"  - Number of edges: {G.number_of_edges()}")
print(f"  - Graph density: {nx.density(G):.3f}")
```

```python
# ----------------------------------------------------------------------------
# Initialize Graph Factorization Algorithm
# ----------------------------------------------------------------------------
gf = GraphFactorization(
    d=2,                    # Embedding dimension (2D for easy visualization)
    max_iter=10000,         # Maximum number of training iterations
    eta=1e-4,               # Learning rate (controls step size in optimization)
    regu=1.0                # Regularization strength (prevents overfitting)
)

# Train the model to learn node embeddings
gf.learn_embedding(G)
print()

# Extract the learned embeddings (each node gets a d-dimensional vector)
gf_embeddings = gf.get_embedding()

print(f"Embedding shape: {gf_embeddings.shape}")
print(f"Sample embedding for node 0: {gf_embeddings[0]}")
print(f"Sample embedding for node 10: {gf_embeddings[10]}")
```

```python
# ----------------------------------------------------------------------------
# Visualization
# ----------------------------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Original graph structure
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, ax=axes[0], node_color='lightblue', 
        node_size=500, with_labels=True, font_size=8,
        edge_color='gray', width=1.5)
axes[0].set_title('Original Barbell Graph Structure', fontsize=14, fontweight='bold')
axes[0].axis('off')

# Plot 2: Learned embeddings in 2D space
# Color nodes by their community (left clique, bridge, right clique)
colors = []
for node in range(G.number_of_nodes()):
    if node < 10:
        colors.append('red')  # Left clique
    elif node < 14:
        colors.append('green')  # Bridge
    else:
        colors.append('blue')  # Right clique

axes[1].scatter(gf_embeddings[:, 0], gf_embeddings[:, 1], 
                c=colors, s=200, alpha=0.7, edgecolors='black', linewidth=1.5)

# Add node labels
for i in range(len(gf_embeddings)):
    axes[1].annotate(str(i), (gf_embeddings[i, 0], gf_embeddings[i, 1]),
                    ha='center', va='center', fontsize=8, fontweight='bold')

axes[1].set_xlabel('Embedding Dimension 1', fontsize=12)
axes[1].set_ylabel('Embedding Dimension 2', fontsize=12)
axes[1].set_title('Learned 2D Node Embeddings', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].legend(['Left Clique', 'Bridge', 'Right Clique'], loc='best')

plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------------------------------------
# Analysis: Reconstruction Quality
# ----------------------------------------------------------------------------
A_original = nx.to_numpy_array(G)
A_reconstructed = gf.get_reconstructed_adj()

reconstruction_error = np.sum((A_original - A_reconstructed) ** 2)

print(f"Total reconstruction error: {reconstruction_error:.4f}")
print(f"Average error per edge: {reconstruction_error / G.number_of_edges():.4f}")
print(f"Frobenius norm of difference: {np.linalg.norm(A_original - A_reconstructed, 'fro'):.4f}")
```

## HOPE (High-Order Proximity Preserved Embedding)

```python
# HOPE is an advanced embedding method that preserves high-order proximity in graphs.
# Unlike GF which focuses on direct connections, HOPE captures both direct and 
# indirect relationships between nodes.

# Key features:
# - Preserves asymmetric transitivity (important for directed graphs)
# - Captures higher-order proximity (paths of length > 1)
# - More scalable than other matrix factorization methods
```

```python
from scipy.linalg import svd
from scipy.sparse import csr_matrix
from sklearn.decomposition import PCA

class HOPE:
    """
    HOPE: High-Order Proximity Preserved Embedding
    
    HOPE preserves high-order proximity in graphs by factorizing a similarity
    matrix (like Katz Index) that captures both direct and indirect connections.
    
    The algorithm factorizes S ≈ U * V^T where:
    - S is a proximity matrix (e.g., Katz similarity)
    - U and V are the source and target embedding matrices
    
    Parameters:
    -----------
    d : int
        Embedding dimension (actual embedding will be d/2 for source and target)
    beta : float
        Decay factor for high-order proximity (controls path length importance)
        Smaller beta = focus on shorter paths, Larger beta = consider longer paths
    """
    
    def __init__(self, d=4, beta=0.01):
        self.d = d
        self.beta = beta
        self.embedding = None
        self._node_list = None
        
    def _compute_katz_similarity(self, A, beta):
        """
        Compute Katz similarity matrix: S = (I - beta*A)^-1 - I
        
        Katz similarity measures node proximity by summing over all paths
        between nodes, with exponentially decaying weights for longer paths.
        
        Parameters:
        -----------
        A : numpy.ndarray
            Adjacency matrix
        beta : float
            Decay factor (must be < 1/λ_max where λ_max is largest eigenvalue)
            
        Returns:
        --------
        S : numpy.ndarray
            Katz similarity matrix
        """
        n = A.shape[0]
        I = np.eye(n)
        
        # Compute S = (I - beta*A)^-1 - I
        # This represents: sum_{k=1}^∞ beta^k * A^k
        try:
            S = np.linalg.inv(I - beta * A) - I
        except np.linalg.LinAlgError:
            # If matrix is singular, use pseudo-inverse
            print("Warning: Using pseudo-inverse due to singularity")
            S = np.linalg.pinv(I - beta * A) - I
            
        return S
    
    def _svd_factorization(self, S, d):
        """
        Perform SVD factorization: S ≈ U * Σ * V^T
        
        We use truncated SVD to get low-rank approximation:
        S ≈ U_d * sqrt(Σ_d) * sqrt(Σ_d) * V_d^T
        
        Parameters:
        -----------
        S : numpy.ndarray
            Similarity matrix to factorize
        d : int
            Number of dimensions to keep
            
        Returns:
        --------
        U_embedding : numpy.ndarray
            Source node embeddings (n x d/2)
        V_embedding : numpy.ndarray
            Target node embeddings (n x d/2)
        """
        # Perform full SVD
        U, sigma, VT = svd(S, full_matrices=False)
        
        # Keep top d/2 dimensions (we'll concatenate U and V later)
        k = d // 2
        U_k = U[:, :k]
        sigma_k = sigma[:k]
        V_k = VT[:k, :].T
        
        # Scale by square root of singular values
        sqrt_sigma = np.diag(np.sqrt(sigma_k))
        
        U_embedding = U_k @ sqrt_sigma
        V_embedding = V_k @ sqrt_sigma
        
        return U_embedding, V_embedding
    
    def learn_embedding(self, graph):
        """
        Learn embeddings that preserve high-order proximity.
        
        Steps:
        1. Compute Katz similarity matrix (captures high-order proximity)
        2. Factorize similarity matrix using SVD
        3. Concatenate source and target embeddings
        
        Parameters:
        -----------
        graph : networkx.Graph
            Input graph
        """
        print(f"Learning HOPE embeddings with d={self.d}, beta={self.beta}")
        
        # Get adjacency matrix
        self._node_list = list(graph.nodes())
        A = nx.to_numpy_array(graph, nodelist=self._node_list)
        n = A.shape[0]
        
        # Check if beta is valid (should be < 1/max_eigenvalue)
        eigenvalues = np.linalg.eigvals(A)
        max_eigenvalue = np.max(np.abs(eigenvalues))
        max_beta = 1.0 / (max_eigenvalue + 1e-6)
        
        if self.beta >= max_beta:
            print(f"Warning: beta={self.beta} is too large (max: {max_beta:.6f})")
            print(f"         Reducing beta to {max_beta * 0.9:.6f}")
            self.beta = max_beta * 0.9
        
        print("Step 1: Computing Katz similarity matrix...")
        S = self._compute_katz_similarity(A, self.beta)
        print(f"        Similarity matrix computed: shape {S.shape}")
        
        print("Step 2: Performing SVD factorization...")
        U_embedding, V_embedding = self._svd_factorization(S, self.d)
        print(f"        Source embeddings: {U_embedding.shape}")
        print(f"        Target embeddings: {V_embedding.shape}")
        
        print("Step 3: Concatenating source and target embeddings...")
        # Final embedding: concatenate source and target representations
        self.embedding = np.hstack([U_embedding, V_embedding])
        print(f"        Final embedding shape: {self.embedding.shape}")
        
        print("HOPE embedding learning complete!")
        
    def get_embedding(self):
        """
        Get the learned node embeddings.
        
        Returns:
        --------
        numpy.ndarray
            Node embedding matrix of shape (n_nodes, d)
        """
        if self.embedding is None:
            raise ValueError("Model has not been trained yet. Call learn_embedding() first.")
        return self.embedding
    
    def get_reconstructed_adj(self):
        """
        Reconstruct the similarity matrix from embeddings.
        
        Returns:
        --------
        numpy.ndarray
            Reconstructed similarity matrix
        """
        if self.embedding is None:
            raise ValueError("Model has not been trained yet. Call learn_embedding() first.")
        
        # Split back into source and target
        k = self.d // 2
        U = self.embedding[:, :k]
        V = self.embedding[:, k:]
        
        # Reconstruct: S_hat = U * V^T
        return U @ V.T
```

```python
# ============================================================================
# DEMONSTRATION: Compare Graph Factorization vs HOPE
# ============================================================================

# Create a barbell graph
G = nx.barbell_graph(m1=10, m2=4)

print(f"\nGraph Statistics:")
print(f"  - Number of nodes: {G.number_of_nodes()}")
print(f"  - Number of edges: {G.number_of_edges()}")
print(f"  - Graph density: {nx.density(G):.3f}")
print(f"  - Average clustering: {nx.average_clustering(G):.3f}")
```

```python
# ============================================================================
# METHOD 1: Graph Factorization (from previous implementation)
# ============================================================================
print("\n" + "=" * 80)
print("METHOD 1: GRAPH FACTORIZATION")
print("=" * 80)

class GraphFactorization:
    def __init__(self, d=128, max_iter=5000, eta=1e-4, regu=1.0):
        self.d = d
        self.max_iter = max_iter
        self.eta = eta
        self.regu = regu
        self.embedding = None
        
    def learn_embedding(self, graph):
        A = nx.to_numpy_array(graph)
        n = A.shape[0]
        np.random.seed(42)
        U = np.random.randn(n, self.d) * 0.01
        
        for iteration in range(self.max_iter):
            A_hat = U @ U.T
            diff = A - A_hat
            loss = np.sum(diff ** 2) + self.regu * np.sum(U ** 2)
            grad = -2 * (diff @ U) + 2 * self.regu * U
            U -= self.eta * grad
            
            if (iteration + 1) % 1000 == 0:
                print(f"  Iteration {iteration + 1}/{self.max_iter} - Loss: {loss:.4f}")
        
        self.embedding = U
    
    def get_embedding(self):
        return self.embedding

gf = GraphFactorization(d=4, max_iter=5000, eta=1e-4, regu=1.0)
gf.learn_embedding(G)
gf_embeddings = gf.get_embedding()

print(f"\nGraph Factorization Results:")
print(f"  Embedding shape: {gf_embeddings.shape}")
print(f"  Sample node 0: {gf_embeddings[0]}")
```

```python
# ============================================================================
# METHOD 2: HOPE
# ============================================================================

hope = HOPE(d=4, beta=0.01)
hope.learn_embedding(G)
hope_embeddings = hope.get_embedding()

print(f"\nHOPE Results:")
print(f"  Embedding shape: {hope_embeddings.shape}")
print(f"  Sample node 0: {hope_embeddings[0]}")
```

```python
# ============================================================================
# VISUALIZATION: Compare Both Methods
# ============================================================================
print("\n" + "=" * 80)
print("VISUALIZATION")
print("=" * 80)

# Reduce to 2D for visualization
pca = PCA(n_components=2)
gf_2d = pca.fit_transform(gf_embeddings)
hope_2d = pca.fit_transform(hope_embeddings)

# Define colors by community
colors = ['#FF6B6B']*10 + ['#4ECDC4']*4 + ['#45B7D1']*10
labels = ['Left Clique'] * 10 + ['Bridge'] * 4 + ['Right Clique'] * 10

fig = plt.figure(figsize=(18, 5))

# Plot 1: Original Graph
ax1 = plt.subplot(131)
pos = nx.spring_layout(G, seed=42)
nx.draw_networkx_nodes(G, pos, node_color=colors, node_size=400, alpha=0.8, ax=ax1)
nx.draw_networkx_edges(G, pos, alpha=0.3, width=1.5, ax=ax1)
nx.draw_networkx_labels(G, pos, font_size=8, font_weight='bold', ax=ax1)
ax1.set_title('Original Barbell Graph', fontsize=14, fontweight='bold', pad=20)
ax1.axis('off')

# Plot 2: Graph Factorization Embeddings
ax2 = plt.subplot(132)
scatter1 = ax2.scatter(gf_2d[:, 0], gf_2d[:, 1], c=colors, s=300, 
                       alpha=0.7, edgecolors='black', linewidth=2)
for i in range(len(gf_2d)):
    ax2.annotate(str(i), (gf_2d[i, 0], gf_2d[i, 1]),
                ha='center', va='center', fontsize=9, fontweight='bold')
ax2.set_xlabel('PC 1', fontsize=11)
ax2.set_ylabel('PC 2', fontsize=11)
ax2.set_title('Graph Factorization\n(First-Order Proximity)', 
              fontsize=14, fontweight='bold', pad=20)
ax2.grid(True, alpha=0.3, linestyle='--')

# Plot 3: HOPE Embeddings
ax3 = plt.subplot(133)
scatter2 = ax3.scatter(hope_2d[:, 0], hope_2d[:, 1], c=colors, s=300,
                       alpha=0.7, edgecolors='black', linewidth=2)
for i in range(len(hope_2d)):
    ax3.annotate(str(i), (hope_2d[i, 0], hope_2d[i, 1]),
                ha='center', va='center', fontsize=9, fontweight='bold')
ax3.set_xlabel('PC 1', fontsize=11)
ax3.set_ylabel('PC 2', fontsize=11)
ax3.set_title('HOPE\n(High-Order Proximity)', 
              fontsize=14, fontweight='bold', pad=20)
ax3.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```

```python
# ============================================================================
# ANALYSIS: Key Differences
# ============================================================================
print("\n1. GRAPH FACTORIZATION:")
print("   • Preserves FIRST-ORDER proximity (direct edges)")
print("   • Factorizes adjacency matrix: A ≈ U * U^T")
print("   • Good for: Local community detection")
print("   • Fast and simple optimization")

print("\n2. HOPE:")
print("   • Preserves HIGH-ORDER proximity (paths of all lengths)")
print("   • Factorizes Katz similarity: S ≈ U * V^T")
print("   • Good for: Long-range dependencies, global structure")
print("   • Captures transitive relationships")

print("\n3. WHEN TO USE WHICH:")
print("   • Use GF when: Direct connections matter most")
print("   • Use HOPE when: Indirect relationships important (e.g., social networks)")
print("   • Use HOPE for: Asymmetric/directed graphs")
print("   • Use GF for: Faster computation, simpler interpretation")

# Compute similarity between nodes in the same vs different cliques
def compute_avg_similarity(embeddings, group1_indices, group2_indices):
    """Compute average cosine similarity between two groups"""
    from sklearn.metrics.pairwise import cosine_similarity
    
    group1_emb = embeddings[group1_indices]
    group2_emb = embeddings[group2_indices]
    
    similarities = cosine_similarity(group1_emb, group2_emb)
    return np.mean(similarities)

left_clique = list(range(10))
right_clique = list(range(14, 24))
bridge = list(range(10, 14))

print("\n" + "=" * 80)
print("EMBEDDING QUALITY ANALYSIS")
print("=" * 80)

# Within-community similarity
gf_within_left = compute_avg_similarity(gf_embeddings, left_clique, left_clique)
hope_within_left = compute_avg_similarity(hope_embeddings, left_clique, left_clique)

# Between-community similarity
gf_between = compute_avg_similarity(gf_embeddings, left_clique, right_clique)
hope_between = compute_avg_similarity(hope_embeddings, left_clique, right_clique)

print(f"\nWithin Left Clique Similarity:")
print(f"  GF:   {gf_within_left:.4f}")
print(f"  HOPE: {hope_within_left:.4f}")

print(f"\nBetween Cliques Similarity:")
print(f"  GF:   {gf_between:.4f}")
print(f"  HOPE: {hope_between:.4f}")

print(f"\nSeparation Ratio (Within/Between - higher is better):")
print(f"  GF:   {gf_within_left/gf_between:.4f}")
print(f"  HOPE: {hope_within_left/hope_between:.4f}")
```
