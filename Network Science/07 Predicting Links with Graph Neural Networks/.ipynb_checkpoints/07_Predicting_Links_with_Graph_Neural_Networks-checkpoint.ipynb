# Predicting Links with Graph Neural Networks

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import torch
import torch.nn.functional as F
from torch.nn import Conv1d, MaxPool1d, Linear, Dropout, BCEWithLogitsLoss

# PyTorch Geometric imports
import torch_geometric.transforms as T
from torch_geometric.datasets import Planetoid
from torch_geometric.nn import GCNConv, VGAE, aggr
from torch_geometric.transforms import RandomLinkSplit
from torch_geometric.data import Data
from torch_geometric.loader import DataLoader
from torch_geometric.utils import k_hop_subgraph, to_scipy_sparse_matrix

# Sklearn and scipy imports
from sklearn.metrics import roc_auc_score, average_precision_score
from scipy.sparse.csgraph import shortest_path

# Set random seeds for reproducibility
np.random.seed(0)
torch.manual_seed(0)
torch.cuda.manual_seed(0)
torch.cuda.manual_seed_all(0)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
```

## Traditional Link Prediction (Using linkpred library)

```python
import linkpred as lp

# Create a sample network (Karate Club graph)
G = nx.karate_club_graph()
nx.draw(G, with_labels=True)
plt.title("Karate Club Graph")
plt.show()
```

### Computing the desired unsupervied predictors

``linkpred`` offers a wide number of unsupervied predictors organized into four families:
- **Neighborhood:**
    - *AdamicAdar*, AssociationStrength, *CommonNeighbours*, Cosine, DegreeProduct, *Jaccard*, MaxOverlap, MinOverlap, NMeasure, Pearson,ResourceAllocation
- **Paths:**
    - GraphDistance, *Katz*
- **Ranking:**
    - *SimRank*, RootedPageRank
- **Miscellanea:**
    - Community, Copy, *Random*



 
**Neighborhood**

The first family of predictors is the one that tries to relate neighborhoods of node pairs with the likelihood of observing a tie among them.
The question such predictors try to answer is: How many friend we have to share in order to become friends?

### Common Neighbors Predictor

```python
# Assumption: The more mutual friends two nodes share, the more likely they 
# will form a connection
print("\n--- Common Neighbors ---")
cn = lp.predictors.CommonNeighbours(G, excluded=G.edges())
cn_results = cn.predict()
top_cn = cn_results.top(5)
for edge, score in top_cn.items():
    print(f"Edge {edge}: Score = {score}")
```

### Jaccard Coefficient Predictor

```python
# Assumption: Nodes with similar neighborhood profiles are likely to connect
# Jaccard = |N(u) ∩ N(v)| / |N(u) ∪ N(v)|
print("\n--- Jaccard Coefficient ---")
jc = lp.predictors.Jaccard(G, excluded=G.edges())
jc_results = jc.predict()
top_jc = jc_results.top(5)
for edge, score in top_jc.items():
    print(f"Edge {edge}: Score = {score:.4f}")
```

### Adamic-Adar Predictor

```python
# Assumption: Common neighbors with fewer connections are more important
# AA = Σ(1 / log|N(z)|) for all common neighbors z
print("\n--- Adamic-Adar ---")
aa = lp.predictors.AdamicAdar(G, excluded=G.edges())
aa_results = aa.predict()
top_aa = aa_results.top(5)
for edge, score in top_aa.items():
    print(f"Edge {edge}: Score = {score:.4f}")
```

## Variational Graph Autoencoder (VGAE)

### Data Preparation

```python
# Set device (GPU if available, otherwise CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\nUsing device: {device}")

# Create transform pipeline:
# - NormalizeFeatures: Normalize node features
# - ToDevice: Move data to GPU/CPU
# - RandomLinkSplit: Split edges into train/val/test sets
transform = T.Compose([
    T.NormalizeFeatures(),
    T.ToDevice(device),
    T.RandomLinkSplit(
        num_val=0.05,           # 5% edges for validation
        num_test=0.1,            # 10% edges for testing
        is_undirected=True,      # Treat graph as undirected
        split_labels=True,       # Create labels for positive/negative samples
        add_negative_train_samples=False  # VGAE handles negative sampling
    )
])

# Load Cora dataset (citation network)
dataset = Planetoid('.', name='Cora', transform=transform)
train_data, val_data, test_data = dataset[0]

print(f"\nDataset: {dataset}")
print(f"Number of features: {dataset.num_features}")
print(f"Number of classes: {dataset.num_classes}")
```

### Encoder Architecture

```python
class Encoder(torch.nn.Module):
    """
    VGAE Encoder: Maps node features to latent space distributions
    
    Architecture:
    - Shared GCN layer: Extracts high-level representations
    - Mean layer (μ): Outputs mean of latent distribution
    - LogStd layer (log σ): Outputs log standard deviation of latent distribution
    """
    def __init__(self, dim_in, dim_out):
        super().__init__()
        # Shared convolutional layer (input -> 2*hidden_dim)
        self.conv1 = GCNConv(dim_in, 2 * dim_out)
        
        # Mean layer for variational distribution
        self.conv_mu = GCNConv(2 * dim_out, dim_out)
        
        # Log standard deviation layer for variational distribution
        self.conv_logstd = GCNConv(2 * dim_out, dim_out)
    
    def forward(self, x, edge_index):
        """
        Forward pass through encoder
        
        Args:
            x: Node features [num_nodes, num_features]
            edge_index: Edge connectivity [2, num_edges]
            
        Returns:
            mu: Mean of latent distribution [num_nodes, dim_out]
            logstd: Log std of latent distribution [num_nodes, dim_out]
        """
        # Apply shared layer with ReLU activation
        x = self.conv1(x, edge_index).relu()
        
        # Return mean and log standard deviation
        return self.conv_mu(x, edge_index), self.conv_logstd(x, edge_index)
```

### Model Initialization

```python
# Initialize VGAE with custom encoder
# The decoder is implicitly an inner product: A_hat = sigmoid(Z @ Z^T)
model = VGAE(Encoder(dataset.num_features, 16)).to(device)

# Use Adam optimizer with learning rate 0.01
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

print(f"\nModel: {model}")
```

### Training and Evaluation Functions 

```python
def train_vgae():
    """
    Train VGAE for one epoch
    
    Loss = Reconstruction Loss + KL Divergence
    - Reconstruction: Binary cross-entropy between predicted and actual edges
    - KL Divergence: Regularization term to keep latent distribution close to N(0,1)
    """
    model.train()
    optimizer.zero_grad()
    
    # Encode: Sample latent representations from learned distributions
    z = model.encode(train_data.x, train_data.edge_index)
    
    # Calculate reconstruction loss (decoder is implicit in recon_loss)
    recon_loss = model.recon_loss(z, train_data.pos_edge_label_index)
    
    # Calculate KL divergence (regularization)
    kl_loss = (1 / train_data.num_nodes) * model.kl_loss()
    
    # Total ELBO loss
    loss = recon_loss + kl_loss
    
    loss.backward()
    optimizer.step()
    
    return float(loss)

@torch.no_grad()
def test_vgae(data):
    """
    Evaluate VGAE on validation/test data
    
    Returns:
        auc: Area Under ROC Curve
        ap: Average Precision
    """
    model.eval()
    z = model.encode(data.x, data.edge_index)
    return model.test(z, data.pos_edge_label_index, data.neg_edge_label_index)
```

### Training Loop

```python
print("\nTraining VGAE...")
for epoch in range(301):
    loss = train_vgae()
    val_auc, val_ap = test_vgae(val_data)
    
    if epoch % 50 == 0:
        print(f'Epoch {epoch:>3} | Loss: {loss:.4f} | Val AUC: {val_auc:.4f} | Val AP: {val_ap:.4f}')

# Final evaluation on test set
test_auc_vgae, test_ap_vgae = test_vgae(test_data)
print(f'\n{"="*60}')
print(f'VGAE Test Results | AUC: {test_auc_vgae:.4f} | AP: {test_ap_vgae:.4f}')
```

### Generate Predicted Adjacency Matrix

```python
# Encode test data to get latent representations
z = model.encode(test_data.x, test_data.edge_index)

# Compute predicted adjacency matrix: A_hat = sigmoid(Z @ Z^T)
Ahat = torch.sigmoid(z @ z.T)

print(f"\nPredicted Adjacency Matrix shape: {Ahat.shape}")
print(f"Sample values:\n{Ahat[:5, :5]}")
```

## SEAL (Subgraph Extraction and Labeling)

### Data Preparation

```python
# Load Cora dataset with link-level split
transform_seal = RandomLinkSplit(
    num_val=0.05,
    num_test=0.1,
    is_undirected=True,
    split_labels=True
)

dataset_seal = Planetoid('.', name='Cora', transform=transform_seal)
train_data_seal, val_data_seal, test_data_seal = dataset_seal[0]

print(f"\nDataset prepared for SEAL")
print(f"Train edges: {train_data_seal.pos_edge_label_index.shape[1]}")
print(f"Val edges: {val_data_seal.pos_edge_label_index.shape[1]}")
print(f"Test edges: {test_data_seal.pos_edge_label_index.shape[1]}")
```

### SEAL Processing Function

```python
def seal_processing(dataset, edge_label_index, y):
    """
    Extract and label enclosing subgraphs for each edge
    
    SEAL Pipeline:
    1. Extract k-hop subgraph around target edge
    2. Remove target edge from subgraph
    3. Apply Double-Radius Node Labeling (DRNL)
    4. Concatenate features with one-hot encoded labels
    
    Args:
        dataset: Graph dataset
        edge_label_index: Edges to process [2, num_edges]
        y: Label (1 for positive, 0 for negative edges)
        
    Returns:
        List of Data objects containing subgraphs
    """
    data_list = []
    
    # Process each edge pair
    for src, dst in edge_label_index.t().tolist():
        
        # Step 1: Extract 2-hop subgraph around source and destination nodes
        sub_nodes, sub_edge_index, mapping, _ = k_hop_subgraph(
            [src, dst],          # Target nodes
            2,                   # Number of hops
            dataset.edge_index,  # Full graph edges
            relabel_nodes=True   # Relabel nodes to [0, num_nodes-1]
        )
        
        # Get new indices of source and destination in subgraph
        src, dst = mapping.tolist()
        
        # Step 2: Remove target edge from subgraph (to avoid data leakage)
        mask1 = (sub_edge_index[0] != src) | (sub_edge_index[1] != dst)
        mask2 = (sub_edge_index[0] != dst) | (sub_edge_index[1] != src)
        sub_edge_index = sub_edge_index[:, mask1 & mask2]
        
        # Step 3: Double-Radius Node Labeling (DRNL)
        # Ensure src < dst for consistency
        src, dst = (dst, src) if src > dst else (src, dst)
        
        # Convert to scipy sparse matrix for distance calculations
        adj = to_scipy_sparse_matrix(
            sub_edge_index,
            num_nodes=sub_nodes.size(0)
        ).tocsr()
        
        # Create adjacency matrix without source node
        idx = list(range(src)) + list(range(src + 1, adj.shape[0]))
        adj_wo_src = adj[idx, :][:, idx]
        
        # Create adjacency matrix without destination node
        idx = list(range(dst)) + list(range(dst + 1, adj.shape[0]))
        adj_wo_dst = adj[idx, :][:, idx]
        
        # Calculate shortest path distances from all nodes to source
        # Note: In adj_wo_dst, the dst node has been removed, so src index may shift
        d_src = shortest_path(
            adj_wo_dst,
            directed=False,
            unweighted=True,
            indices=src if src < dst else src - 1  # Adjust index if src > dst
        )
        d_src = np.insert(d_src, dst, 0, axis=0)  # Distance from src to dst is 0
        d_src = torch.from_numpy(d_src)
        
        # Calculate shortest path distances from all nodes to destination
        # Note: In adj_wo_src, the src node has been removed, so dst index shifts down
        d_dst = shortest_path(
            adj_wo_src,
            directed=False,
            unweighted=True,
            indices=dst - 1  # dst index is now one less since src was removed
        )
        d_dst = np.insert(d_dst, src, 0, axis=0)  # Distance from dst to src is 0
        d_dst = torch.from_numpy(d_dst)
        
        # Calculate DRNL labels using the formula:
        # z = 1 + min(d_src, d_dst) + (d_src + d_dst)//2 * ((d_src + d_dst)//2 + (d_src + d_dst)%2 - 1)
        dist = d_src + d_dst
        z = 1 + torch.min(d_src, d_dst) + dist // 2 * (dist // 2 + dist % 2 - 1)
        
        # Set labels for source, destination, and unreachable nodes
        z[src], z[dst], z[torch.isnan(z)] = 1., 1., 0.
        z = z.to(torch.long)
        
        # Step 4: Concatenate node features with one-hot encoded DRNL labels
        node_labels = F.one_hot(z, num_classes=200).to(torch.float)
        node_emb = dataset.x[sub_nodes]  # Original node features
        node_x = torch.cat([node_emb, node_labels], dim=1)
        
        # Create Data object for this subgraph
        data = Data(
            x=node_x,              # Combined features and labels
            z=z,                    # DRNL labels
            edge_index=sub_edge_index,  # Subgraph edges
            y=y                     # Edge label (0 or 1)
        )
        data_list.append(data)
    
    return data_list
```

### Extract Enclosing Subgraphs

```python
print("\nExtracting enclosing subgraphs...")

# Process positive and negative edges separately
train_pos_list = seal_processing(train_data_seal, train_data_seal.pos_edge_label_index, 1)
train_neg_list = seal_processing(train_data_seal, train_data_seal.neg_edge_label_index, 0)

val_pos_list = seal_processing(val_data_seal, val_data_seal.pos_edge_label_index, 1)
val_neg_list = seal_processing(val_data_seal, val_data_seal.neg_edge_label_index, 0)

test_pos_list = seal_processing(test_data_seal, test_data_seal.pos_edge_label_index, 1)
test_neg_list = seal_processing(test_data_seal, test_data_seal.neg_edge_label_index, 0)

# Merge positive and negative examples
train_dataset = train_pos_list + train_neg_list
val_dataset = val_pos_list + val_neg_list
test_dataset = test_pos_list + test_neg_list

print(f"Train subgraphs: {len(train_dataset)}")
print(f"Val subgraphs: {len(val_dataset)}")
print(f"Test subgraphs: {len(test_dataset)}")

# Create data loaders for batch processing
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)
test_loader = DataLoader(test_dataset, batch_size=32)
```

### DGCNN Architecture

```python
class DGCNN(torch.nn.Module):
    """
    Deep Graph Convolutional Neural Network for Link Prediction
    
    Architecture:
    1. GCN layers: Extract graph features from subgraphs
    2. SortPooling: Global pooling that preserves node ordering
    3. 1D Convolutions: Process sorted node representations
    4. Dense layers: Final classification
    
    Args:
        dim_in: Input feature dimension
        k: Number of nodes to keep after SortPooling
    """
    def __init__(self, dim_in, k=30):
        super().__init__()
        
        # 1. Graph Convolutional Layers
        self.gcn1 = GCNConv(dim_in, 32)
        self.gcn2 = GCNConv(32, 32)
        self.gcn3 = GCNConv(32, 32)
        self.gcn4 = GCNConv(32, 1)
        
        # 2. Global Sort Pooling
        # Keeps top-k nodes based on their features
        self.global_pool = aggr.SortAggregation(k=k)
        
        # 3. Traditional Convolutional Layers (1D)
        # These process the sorted node representations
        self.conv1 = Conv1d(1, 16, kernel_size=97, stride=97)
        self.conv2 = Conv1d(16, 32, kernel_size=5, stride=1)
        self.maxpool = MaxPool1d(kernel_size=2, stride=2)
        
        # 4. Dense Layers for Classification
        self.linear1 = Linear(352, 128)
        self.dropout = Dropout(0.5)
        self.linear2 = Linear(128, 1)
    
    def forward(self, x, edge_index, batch):
        """
        Forward pass through DGCNN
        
        Args:
            x: Node features [num_nodes, num_features]
            edge_index: Edge connectivity [2, num_edges]
            batch: Batch assignment [num_nodes]
            
        Returns:
            predictions: Link existence probability [batch_size, 1]
        """
        # Step 1: Apply GCN layers with tanh activation
        h1 = self.gcn1(x, edge_index).tanh()
        h2 = self.gcn2(h1, edge_index).tanh()
        h3 = self.gcn3(h2, edge_index).tanh()
        h4 = self.gcn4(h3, edge_index).tanh()
        
        # Concatenate all layer outputs for richer representations
        h = torch.cat([h1, h2, h3, h4], dim=-1)
        
        # Step 2: Apply global sort pooling
        h = self.global_pool(h, batch)
        
        # Step 3: Reshape for 1D convolutions [batch, channels, length]
        h = h.view(h.size(0), 1, h.size(-1))
        
        # Apply convolutional layers
        h = self.conv1(h).relu()
        h = self.maxpool(h)
        h = self.conv2(h).relu()
        
        # Step 4: Flatten and apply dense layers
        h = h.view(h.size(0), -1)
        h = self.linear1(h).relu()
        h = self.dropout(h)
        h = self.linear2(h).sigmoid()
        
        return h
```

### Model Training

```python
# Initialize model
model_seal = DGCNN(train_dataset[0].num_features, k=30).to(device)
optimizer_seal = torch.optim.Adam(params=model_seal.parameters(), lr=0.0001)
criterion = BCEWithLogitsLoss()

print(f"\nDGCNN Model: {model_seal}")

def train_seal():
    """Train DGCNN for one epoch"""
    model_seal.train()
    total_loss = 0
    
    for data in train_loader:
        data = data.to(device)
        optimizer_seal.zero_grad()
        
        # Forward pass
        out = model_seal(data.x, data.edge_index, data.batch)
        
        # Calculate loss
        loss = criterion(out.view(-1), data.y.to(torch.float))
        
        # Backward pass
        loss.backward()
        optimizer_seal.step()
        
        total_loss += float(loss) * data.num_graphs
    
    return total_loss / len(train_dataset)

@torch.no_grad()
def test_seal(loader):
    """Evaluate DGCNN on validation/test data"""
    model_seal.eval()
    y_pred, y_true = [], []
    
    for data in loader:
        data = data.to(device)
        out = model_seal(data.x, data.edge_index, data.batch)
        
        y_pred.append(out.view(-1).cpu())
        y_true.append(data.y.view(-1).cpu().to(torch.float))
    
    # Calculate metrics
    auc = roc_auc_score(torch.cat(y_true), torch.cat(y_pred))
    ap = average_precision_score(torch.cat(y_true), torch.cat(y_pred))
    
    return auc, ap

# Training loop
print("\nTraining DGCNN (SEAL)...")
for epoch in range(31):
    loss = train_seal()
    val_auc, val_ap = test_seal(val_loader)
    
    if epoch % 5 == 0:
        print(f'Epoch {epoch:>2} | Loss: {loss:.4f} | Val AUC: {val_auc:.4f} | Val AP: {val_ap:.4f}')

# Final evaluation
test_auc_seal, test_ap_seal = test_seal(test_loader)
print(f'\n{"="*60}')
print(f'SEAL Test Results | AUC: {test_auc_seal:.4f} | AP: {test_ap_seal:.4f}')
print(f'{"="*60}')
```

```python
print(f"VGAE  | AUC: {test_auc_vgae:.4f} | AP: {test_ap_vgae:.4f}")
print(f"SEAL  | AUC: {test_auc_seal:.4f} | AP: {test_ap_seal:.4f}")
```
